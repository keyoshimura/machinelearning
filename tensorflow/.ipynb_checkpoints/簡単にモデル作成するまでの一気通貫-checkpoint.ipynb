```python
import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.boston_housing.load_data()
```

```python
%matplotlib inline

'''
ちょっとデータを眺めてみる
'''

import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 10*3
plt.rcParams['figure.figsize'] = [18, 12]
plt.rcParams['font.family'] = ['IPAexGothic']

plt.hist(y_train, bins=20)
plt.xlabel('data_size')
plt.ylabel('house_price(/1,000$)')
plt.show()


plt.plot(x_train[: ,5], y_train, 'o')
plt.xlabel('n_room')
plt.ylabel('house_price(/1,000$)')
```

```python
'''
データの前処理。
「標準化」をしておく。これで説明変数ごとの「スケールの違い」に影響されないようにできる。

'''


x_train_mean = x_train.mean(axis=0)
x_train_std = x_train.std(axis=0)
y_train_mean = y_train.mean()
y_train_std = y_train.std()

x_train = (x_train- x_train_mean) / x_train_std
y_train = (y_train- y_train_mean) / y_train_std
x_test = (x_test- x_train_mean) / x_train_std # testにも「x_train」の値を使う
y_test = (y_test- y_train_mean) / y_train_std


plt.plot(x_train[:, 5], y_train, 'o')
plt.xlabel('n_room(normalized)')
plt.ylabel('house_price(/1,000$)(normalized)')
```

```python
'''
モデルの学習
'''

import tensorflow as tf

# 説明変数用のプレースホルダー
x = tf.placeholder(tf.float32, (None, 13), name='x')
# 正解データ(住宅価格)用のプレースホルダー
y = tf.placeholder(tf.float32, (None, 1), name='y')

# モデルの作成（説明変数を重み w で足し合わせただけの簡単なやつにした）
w = tf.Variable(tf.random_normal((13, 1)))
pred = tf.matmul(x, w)

# 実データと推定値の差の二乗の平均を誤差とする
loss = tf.reduce_mean((y - pred)**2)
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate=0.1
)
train_step = optimizer.minimize(loss)

n_epochs = 200
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(n_epochs):
        # train_step が None を返すので、 _ で受けておく
        train_loss, _ = sess.run(
            [loss, train_step],
            feed_dict={
                x: x_train,
                # y_trainとyの次元を揃えるためにreshapeが必要
                y: y_train.reshape((-1, 1))
            }
        )
        print('step: {}, train_loss: {}'.format(
            step, train_loss
        ))

    # 学習が終わったら、評価用データに対して予測してみる
    pred_ = sess.run(
        pred,
        feed_dict={
            x: x_test
        }
    )
```

```python
'''
予測結果を可視化して、どのくらい「それっぽいの」が出ているかを確認する
'''

plt.plot(x_test[:, 5], pred_, 'o', label='pred')
plt.plot(x_test[:, 5], y_test, 'x', label='true')
plt.legend(loc='upper left')
plt.xlabel('cnt_room')
plt.ylabel('house_price')
```

```python
'''おまけ
ミニバッチ学習関係
'''

# ミニバッチを1個ずつ返却するジェネレータ
import numpy as np

def get_batches(x, y, batch_size):
    n_data = len(x)
    indices = np.arange(n_data)
    np.random.shuffle(indices)
    x_shuffled = x[indices]
    y_shuffled = y[indices]
    
    # 元データからバッチサイズ分のデータを抽出して返却する
    for i in range(0, n_data, batch_size):
        x_batch = x_shuffled[i: i+batch_size]
        y_batch = y_shuffled[i: i+batch_size]
        yield x_batch, y_batch


# ミニバッチを使った学習
BATCH_SIZE = 32
step = 0
n_epochs = 100

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    # n_epochs分だけ回す
    for epoch in range(n_epochs):
        for x_batch, y_batch in get_batches(x_train, y_train, BATCH_SIZE):
            train_loss, _ = sess.run(
                [loss, train_step],
                feed_dict={
                    x: x_batch,
                    y: y_batch.reshape((-1, 1))
                }
            )
            print('step:{0}, train_loss:{1}'.format(step, train_loss)
                 )
            step += 1
            
            pred_ = sess.run(pred,
                             feed_dict={
                                 x: x_test
                             }
                            )
```
